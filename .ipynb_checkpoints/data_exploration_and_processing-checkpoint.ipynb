## Processing of the EMG data

### Notch filter
A notch filter is a band-stop filter with a narrow bandwidth (high quality factor). It rejects a narrow frequency band and leaves the rest of the spectrum little changed.
Here the implementation from [scipy](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.iirnotch.html) is used

```python
from scipy.signal import iirnotch
from scipy import signal
```

```python
# parameters
fs = 500.0  # Sample frequency (Hz)
f0 = 60.0  # Frequency to be removed from signal (Hz)
Q = 35.0  # Quality factor
# Design notch filter
b, a = signal.iirnotch(f0, Q, fs)
```

###  Butterworth filter
Again taken from [scipy](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.butter.html).

```python
N = 4 # The order of the filter.
f0 = 60.0  # Frequency to be removed from signal (Hz)
sos = signal.butter(N, f0, 'lowpass', 1000, output='sos')
```

```python
# Take one EMG signal to process
signal_emg = data.X[N]['EMG1'].to_numpy()
```

```python
# lets apply the two filters on a single sample
# butterworth
filtered = signal.sosfilt(sos, signal_emg)
notch_filtered=signal.lfilter(b, a, filtered)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.6, 0.75])
for s in [signal_emg, notch_filtered]:
    ax.plot(s, marker='o')

plt.title('Initial signal (blue) and filtered signal (orange)')
plt.show()
```
