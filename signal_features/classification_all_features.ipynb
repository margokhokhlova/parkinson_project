```python
data_path = 'C:/Users/khokhlovam/Documents/kotelnikov/data/data_lstm_test_complet.csv'
df =pd.read_csv(data_path, header=None, names=range(1564)) # 
```

```python
# Let's use mean values of all the characteristics, and then let's separate classes as good as we can with those
```

```python
mean_df = pd.DataFrame(columns=['label', 'ID', 'parameter', 'value', ])
for i in range(0,len(df)): # 
    val = df.iloc[[i]].values.tolist()[0]
    valid_indexes = [j for j in range(3,len(val)) if str(val[j]) !='nan']
    valid_val = [val[x] for x in valid_indexes] 
    label =  df.iloc[[i]][0].values[0] 
    ID = df.iloc[[i]][1].values[0] 
    parameter= df.iloc[[i]][2].values[0] 
    mean_df.loc[len(mean_df)]= [label, ID, parameter, np.mean(valid_val)]
    if parameter == 'phase':
        total_wavetrains= len(valid_indexes)
        mean_df.loc[len(mean_df)]= [label, ID, 'total_wavetrains',len(valid_val)]
```

```python
pd.options.display.float_format = '{:.2f}'.format
print(mean_df.to_string())
```

```python
# Let's do Random Classifier on those features
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
```

```python
X = []
y = []
feat_patient = []
for i in range(1,len(mean_df)):
    feat_patient.append(mean_df['value'][i-1])
    if i%10==0:
        X.append(feat_patient)
        feat_patient = []
        y.append(mean_df['label'][i])
        
```

```python
X, y, len(X), len(y)
```

```python
Y = []
for y_v in y:
    if y_v=='control':
        Y.append(0)
    elif y_v == 'ET':
        Y.append(1)
    else:
        Y.append(2)
```

```python
X = np.array(X)
```

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X, Y)
```

```python
importance_scores = classifier.feature_importances_
```

```python
feat_names = ['freq',  'flash',   'bandwidth', 'timeSignal', 'time_sincelast', 'ampl', 'duratSec', 'frac',  'phase', 'total_wavetrains'] 
importance_ranking = sorted(zip(importance_scores, feat_names), reverse=True)
for score, name in importance_ranking:
    print(f"{name}: {score:.4f}")
```
